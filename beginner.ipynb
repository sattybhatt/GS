```
from transformers import pipeline
import pandas as pd
```

```
classifier = pipeline("text-classification",model="SamLowe/roberta-base-go_emotions")
text="I hate you"
outputs=classifier(text)
pd.DataFrame(outputs)
```

**Named Entity Recognition COde below**

```
ner_tagger=pipeline("ner",aggregation_strategy="simple")
text="My name is Sattyam and I work for Microsoft"

outputs=ner_tagger(text)
pd.DataFrame(outputs)
```

**TEXT GENERATION PIPELINE**

```
from transformers import set_seed
set_seed(0)

```

```
generator=pipeline("text-generation")
text="""
Dear Amazon, last week I ordered an Optimus Prime action figure
from your online store in Germany. Unfortunately, when I opened the package,i discovered to my horror that I had been sent an action figure of Megatron
instead
"""
response="I am sorry to hear that your order was mixed up"

prompt="user: "+text.replace("\n"," ")+"Customer service Response: "+response

outputs=generator(prompt,max_length=128)
print(outputs)
```
